Data Dictionary:

1. ID: Unique id of the patient

2. Age : Age of the patient

3. Gender : Gender of the patient

4. TB : Total Bilirubin5.DB : Direct Bilirubin

6. Alkphos: Alkaline Phosphotase

7. Sgpt:  Alamine Aminotransferase

8. Sgot:  Aspartate Aminotransferase

9. TP:  Total Protiens

10. ALB:  Albumin

11. A/G Ratio : Albumin and Globulin Ratio

12. State_Code: Patient state

13. City_Code: Patient city

```python
# Box Plots for Outlier Detection:

numerical_features = ['Age of the patient', 'Total Bilirubin', 'Direct Bilirubin', 'Alkphos Alkaline Phosphotase', 'Sgpt Alamine Aminotransferase', 'Sgot Aspartate Aminotransferase', 'Total Protiens', 'ALB Albumin', 'A/G Ratio Albumin and Globulin Ratio']
for feature in numerical_features:
    sns.boxplot(x='Status', y=feature, data=df_final)
    plt.show()
```

```python
# Visualize the distribution of categorical features such as Gender, State_Code, City_Code.

categorical_features = ['Gender of the patient', 'State_Code', 'City_Code']
for feature in categorical_features:
    sns.countplot(x=feature, hue='Status', data=df_final)
    plt.show()
```

```python
# Bar plot for A/G Ratio Albumin and Globulin Ratio with liver Disease

plt.figure(figsize=(6, 4))
sns.barplot(x='Status', y='A/G Ratio Albumin and Globulin Ratio', data=df_final)
plt.show()
```

```python
#Scatter Plot for Total Bilirubin vs. Direct Bilirubin with Liver Disease

plt.figure(figsize=(6, 4))
sns.scatterplot(x='Total Bilirubin', y='Direct Bilirubin', hue='Status', data=df_final)
plt.title('Scatter Plot for Total Bilirubin vs. Direct Bilirubin with Liver Disease')
plt.show()
```

# 2. Any affect of age group on the occurrence of liver disease.

```python
bins = [0, 20, 30, 40, 50, 60, 70, 80, 90, 100]
labels = ['0-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90', '91-100']
df_new = df_final
df_new['Age_Group'] = pd.cut(df_new['Age of the patient'], bins=bins, labels=labels, right=False)

age_liver_disease = pd.crosstab(df_new['Age_Group'], df_new['Status'])
print("Effect of Age Group on Liver Disease:\n", age_liver_disease)
```

```python
age_liver_disease = pd.crosstab(df_final['Age_Group'], df_final['Status'])

age_liver_disease.plot(kind='bar') 
plt.title('Effect of Age Group on Liver Disease Status')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Status', labels=['Liver Patient', 'Not Liver Patient'], loc='upper right')
plt.show()
```

# 3.Relationship Between Different Factors and Liver Disease:

```python
df_new = df_new.drop(["ID","City_Code","State_Code","Age_Group"],axis=1)
```

```python
df_new.corr()
```

```python
sns.pairplot(df_new, hue='Status')
plt.show()

# Correlation with the target variable
correlation_with_status = df_new.corr()['Status'].sort_values(ascending=False)
print("Correlation with Liver Disease Status:\n", correlation_with_status)
```

```python
# Display correlation matrix
correlation_matrix = df_new.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

# 4. Dataset Balance, Missing Values, and Duplicate Records:

```python
# Check if the dataset is balanced
class_count = df_final['Status'].value_counts()
print(class_count)

sns.countplot(x='Status', data=df_final)
plt.title('Class Distribution of Liver Disease')
plt.xlabel('Liver Disease Status')
plt.ylabel('Count')
plt.show()

print("Hence, the data is unbalanced")
```

```python
# Check for duplicate records
duplicate_records = df_final.duplicated().sum()
print("Number of Duplicate Records:", duplicate_records)
```
