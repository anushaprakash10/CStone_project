9. TP:  Total Protiens

10. ALB:  Albumin

11. A/G Ratio : Albumin and Globulin Ratio

12. State_Code: Patient state

13. City_Code: Patient city

```python
#dropping rows in which both corelated columns have null values
df_final.dropna(subset=['Total Bilirubin', 'Direct Bilirubin'], how='all', inplace=True)
```

```python
df_final[df_final['Total Bilirubin'] == 1][['Total Bilirubin', 'Direct Bilirubin']]
```

```python
df_final[df_final['Total Bilirubin'] == 0.7]['Direct Bilirubin'].unique()
```

```python
med_DB = df_final.groupby('Total Bilirubin')['Direct Bilirubin'].median()
```

```python
def impute_direct_bilirubin(row):
    total_bilirubin = row['Total Bilirubin']
    if pd.isnull(row['Direct Bilirubin']):
        return med_DB.get(total_bilirubin, row['Direct Bilirubin'])
    else:
        return row['Direct Bilirubin']
```

```python
df_final['Direct Bilirubin'] = df_final.apply(impute_direct_bilirubin, axis=1)
```

```python
df_final.isnull().sum()
```

```python
df_final[df_final['Direct Bilirubin'] == 11.8]['Total Bilirubin'].unique()
```

```python
med_TB = df_final.groupby('Direct Bilirubin')['Total Bilirubin'].median()
```

```python
def impute_total_bilirubin(row):
    direct_bilirubin = row['Direct Bilirubin']
    if pd.isnull(row['Total Bilirubin']):
        return med_TB.get(direct_bilirubin, row['Total Bilirubin'])
    else:
        return row['Total Bilirubin']
```

```python
df_final['Total Bilirubin'] = df_final.apply(impute_total_bilirubin, axis=1)
```

```python
mdALB = df_final.groupby('A/G Ratio Albumin and Globulin Ratio')['\xa0ALB Albumin'].median()
```

```python
def impute_ALB(row):
    AG = row['A/G Ratio Albumin and Globulin Ratio']
    if pd.isnull(row['\xa0ALB Albumin']):
        return mdALB.get(AG, row['\xa0ALB Albumin'])
    else:
        return row['\xa0ALB Albumin']
```

```python
df_final['\xa0ALB Albumin'] = df_final.apply(impute_ALB, axis=1)
```

```python
mdAG = df_final.groupby('\xa0ALB Albumin')['A/G Ratio Albumin and Globulin Ratio'].median()
```

```python
def impute_AZ(row):
    ALB = row['\xa0ALB Albumin']
    if pd.isnull(row['A/G Ratio Albumin and Globulin Ratio']):
        return mdAG.get(ALB, row['A/G Ratio Albumin and Globulin Ratio'])
    else:
        return row['A/G Ratio Albumin and Globulin Ratio']
```

```python
df_final['A/G Ratio Albumin and Globulin Ratio'] = df_final.apply(impute_AZ, axis=1)
```

```python
df_final['Total Protiens'].unique()
```

```python
mdP = df_final.groupby('\xa0ALB Albumin')['Total Protiens'].median()
```

```python
def impute_P(row):
    p = row['\xa0ALB Albumin']
    if pd.isnull(row['Total Protiens']):
        return mdP.get(p, row['Total Protiens'])
    else:
        return row['Total Protiens']
```

```python
df_final['Total Protiens'] = df_final.apply(impute_P, axis=1)
```
